BiometricEnrolment Ratio=
Total Biometric Updates / Total Enrolments

i.  Biometric-Enrolment-ration >1 ----------->multiple biometric updates per enrolment
ii. Biometric-Enrolment-ration <1 -----------> new enrolments dominating
iii.Very low ratio                -----------> early-stage Aadhaar penetration
iv. Very high ratio               -----------> mature Aadhaar ecosystem

```python
state_ratio_df = (
    final_df
    .groupby('state')
    .agg(
        biometric_total = (
            'bio_age_5_17', 'sum'
        ),
        biometric_17_plus = (
            'bio_age_17_greater', 'sum'
        ),
        enrolment_total = (
            'enr_age_18_greater', 'sum'
        )
    )
    .reset_index()
)

# Add missing biometric component
state_ratio_df['biometric_total'] = (
    state_ratio_df['biometric_total'] +
    state_ratio_df['biometric_17_plus']
)

# Add full enrolment
state_ratio_df['enrolment_total'] = (
    final_df
    .groupby('state')[['enr_age_0_5','enr_age_5_17','enr_age_18_greater']]
    .sum()
    .sum(axis=1)
    .values
)

# Ratio
state_ratio_df['bio_to_enrol_ratio'] = (
    state_ratio_df['biometric_total'] /
    state_ratio_df['enrolment_total']
)

state_ratio_df.head()

state_ratio_df_sorted = (
    state_ratio_df
    .sort_values('bio_to_enrol_ratio', ascending=False)
)

plt.figure(figsize=(12,6))

plt.bar(
    state_ratio_df_sorted['state'],
    state_ratio_df_sorted['bio_to_enrol_ratio']
)

plt.xticks(rotation=45, ha='right')
plt.xlabel("State")
plt.ylabel("Biometric / Enrolment Ratio")
plt.title("Biometric to Enrolment Contribution Ratio by State")

plt.tight_layout()
plt.show()
```

BiometricEnrolment Ratio(for 17+)=
Total Biometric Updates (17+)/ Total Enrolments(17+)

```python
age_ratio_df = (
    final_df
    .groupby('state')
    .agg(
        bio_17_plus=('bio_age_17_greater', 'sum'),
        enr_18_plus=('enr_age_18_greater', 'sum')
    )
    .reset_index()
)

# -------------------------------
# 3. Compute ratio
# -------------------------------
age_ratio_df['ratio_17_plus'] = (
    age_ratio_df['bio_17_plus'] / age_ratio_df['enr_18_plus']
)

# -------------------------------
# 4. Handle invalid values
# -------------------------------
age_ratio_df.replace([float('inf'), -float('inf')], pd.NA, inplace=True)
age_ratio_df = age_ratio_df.dropna(subset=['ratio_17_plus'])

# -------------------------------
# 5. Sort for visualization
# -------------------------------
age_ratio_sorted = (
    age_ratio_df
    .sort_values('ratio_17_plus', ascending=False)
)

# -------------------------------
# 6. Plot bar chart
# -------------------------------
plt.figure(figsize=(12, 6))

plt.bar(
    age_ratio_sorted['state'],
    age_ratio_sorted['ratio_17_plus']
)

plt.xticks(rotation=45, ha='right')
plt.xlabel("State")
plt.ylabel("Biometric / Enrolment Ratio (Age  18)")
plt.title("State-wise Biometric to Enrolment Ratio (Age  18)")

plt.tight_layout()
plt.show()
```

```python
AGE_GROUPS = {
    "Biometric (517)": "bio_age_5_17",
    "Biometric (18+)": "bio_age_17_greater",

    "Demographic (517)": "demo_age_5_17",
    "Demographic (18+)": "demo_age_17_greater",

    "Enrolment (05)": "enr_age_0_5",
    "Enrolment (517)": "enr_age_5_17",
    "Enrolment (18+)": "enr_age_18_greater",
}
```

```python
STATE_CANONICAL_MAP = {
    # Renames
    "Orissa": "Odisha",

    # Islands
    "Andaman and Nicobar": "Andaman And Nicobar Islands",

    # UT mergers
    "Dadra and Nagar Haveli": "Dadra And Nagar Haveli And Daman And Diu",
    "Daman and Diu": "Dadra And Nagar Haveli And Daman And Diu",

    

    # Missing states (will be NaN geometry)
    # Telangana, Ladakh  CSV has data but GeoJSON has no shapes
}
```

```python
import json
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load GeoJSON
india = gpd.read_file("../Dataset/GeoJSON/india_state_geo.json")

# Extract authoritative state names
geo_states = india["NAME_1"].unique()

geo_lookup = {
    s.lower().replace("&", "and"): s for s in geo_states
}

def normalize_state(name):
    if pd.isna(name):
        return None
    key = name.lower().replace("&", "and")
    key = " ".join(key.split())
    return geo_lookup.get(key, name)
```

```python

india["state_canonical"] = india["NAME_1"].replace(STATE_CANONICAL_MAP)
```

```python
df = pd.read_csv("../Dataset/final_df.csv")

df["state_norm"] = df["state"].apply(normalize_state)

state_df = (
    df
    .groupby("state_norm", as_index=False)
    .sum(numeric_only=True)
)
```

```python
gdf = india.merge(
    state_df,
    left_on="state_canonical",
    right_on="state_norm",
    how="left"
)
```

```python
for title, col in AGE_GROUPS.items():

    fig, ax = plt.subplots(figsize=(10, 12))

    # log-scale to handle skew
    gdf[f"{col}_log"] = np.log1p(gdf[col])

    gdf.plot(
        column=f"{col}_log",
        ax=ax,
        legend=True,
        missing_kwds={
            "color": "lightgrey",
            "label": "No data"
        }
    )

    ax.set_title(f"India Heatmap  {title}", fontsize=14)
    ax.axis("off")

    plt.tight_layout()
    plt.show()
```
